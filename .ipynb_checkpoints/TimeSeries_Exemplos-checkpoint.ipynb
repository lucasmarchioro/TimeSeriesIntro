# 1 Conceito - Correlao entre duas sries

Por vezes,  bem comum que duas sries variem conjuntamente. Ou seja, quando uma tem um incremento, a outra sobe tambm. Nesse caso, o mesmo acontece quando uma das duas sries tem queda.

```python
correlacao = df_item1['value_sales'].corr(df_item2['value_sales'])
print("Correlao de vendas entre as duas sries: ", np.round(correlacao*100), "%" ) 
plt.scatter(df_item1['value_sales'], df_item2['value_sales'])
plt.show()
```

$\rho  = 1 $ Correlao Totalmente Positiva

$\rho  = -1 $ Correlao Totalmente Negativa

$\rho = 0 $ Sem Correlao




Nem toda correlao entre duas sries significa necessariamente que elas so correlatas... https://www.tylervigen.com/spurious-correlations

Por isso, ao ver correlao entre duas sries, no devemos ver apenas os nveis, mas sim os retornos
